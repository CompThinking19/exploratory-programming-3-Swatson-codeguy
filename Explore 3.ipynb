```python
import re

def my_string(prideandp):
    
    if type(prideandp) != str:
        
        raise TypeError ("Not a String")

    results = re.findall ('[A-Za-z]*at\\b', prideandp)
    
    roots = []
    
    for elements in results:
        
        if len(elements) > 3:
             
            roots.append(elements)
            
    return roots
    
source = open('pg1342.txt')


pride = source.read()    
```

```python
my_string(pride)
```
